#### Discretization

```
import pandas as pd
df = pd.read_csv("nba.csv", index_col="Name")

#I am getting all the ids of these rows that match the conditions
lows = df.index[df['Salary'] < 200000].tolist()
meds = df.index[(df['Salary']>=200000) & (df['Salary']<1000000)].tolist()
highs = df.index[df['Salary']>= 1000000].tolist()

#Once I get the row ids, I will now convert them to string values
df.loc[lows,'Salary']='low'
df.loc[meds, 'Salary']='medium'
df.loc[highs, 'Salary']='high'
print(df['Salary'])
```
